```python
#Import necessary libraries
import numpy as np
from keras.layers import Dense, Activation
from keras.models import Sequential
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
dataset = df
X = dataset.iloc[:,0:126]
y = dataset.iloc[:,126:134]
```

Dataset Splitting

Splitting up into Training, Cross Validation, and Test sets are common best practices. This allows you to tune various parameters of the algorithm without making judgements that specifically conform to training data.

```python
# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.08, random_state = 16)
```

Feature Scaling

Feature scaling is a method used to normalize the range of independent variables or features of data. 

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

hidden layer

A hidden layer in an artificial neural network is a layer in between input layers and output layers, where artificial neurons take in a set of weighted inputs and produce an output through an activation function.

```python
#Define number of suitable layers
#Nh = Ns/(α∗ (Ni + No))
'''
    Ni = número de neuronas de entrada. 126
    No = número de neuronas de salida. 8
    Ns = número de muestras en el conjunto de datos de entrenamiento. 2773
    α = un factor de escala arbitrario generalmente 2-10. 5
'''
nh = 2773/(4*(126 + 8))
nh
```

```python
# Initialising the ANN
model = Sequential()
```

Sigmoid function

Sigmoid Function acts as an activation function in machine learning which is used to add non-linearity in a machine learning model, in simple words it decides which value to pass as output and what not to pass.

```python
# Adding the input layer and the first hidden layer
model.add(Dense(250, activation='sigmoid', input_dim = 126))

# Adding the second hidden layer
model.add(Dense(units = 250, activation = 'sigmoid'))

# Adding the third hidden layer
model.add(Dense(units = 250, activation = 'sigmoid'))

# Adding the fourth hidden layer
model.add(Dense(units = 250, activation = 'sigmoid'))

# Adding the fifth hidden layer
model.add(Dense(units = 250, activation = 'sigmoid'))
```

```python
# Adding the output layer

model.add(Dense(units = 8))
```

Adam

Is an adaptive learning rate optimization algorithm that's been designed specifically for training deep neural networks. The algorithms leverages the power of adaptive learning rates methods to find individual learning rates for each parameter.

Loss

The loss function is the function that computes the distance between the current output of the algorithm and the expected output. It's a method to evaluate how your algorithm models the data.

```python
#model.add(Dense(1))
# Compiling the ANN
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
model.fit(X_train, y_train, batch_size = 10, epochs = 100)

y_pred = model.predict(X_test)
```

**Confusion Matrix**

A confusion matrix is a tabular summary of the number of correct and incorrect predictions made by a classifier.

**Accuracy**

Accuracy is one metric for evaluating classification models. Informally, accuracy is the fraction of predictions our model got right.

**Recall**

The recall is the ratio tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives. 

**Precision**

The precision is the ratio tp / (tp + fp) where tp is the number of true positives and fp the number of false positives.
